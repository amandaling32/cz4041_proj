## Fixes the data based on this
https://www.kaggle.com/c/sberbank-russian-housing-market/discussion/34364 

```python
import pandas as pd
```

```python
bad_addresses = pd.read_excel('BAD_ADDRESS_FIX.xlsx')
bad_addresses
```

```python
train = pd.read_csv("C://Users//wongd//OneDrive//Documents//GitHub//machine_learning_project//train.csv")
test = pd.read_csv("C://Users//wongd//OneDrive//Documents//GitHub//machine_learning_project//test.csv")
```

```python
train_bad_addresses = bad_addresses[bad_addresses['id'] <= 30473]
test_bad_addresses = bad_addresses[bad_addresses['id'] > 30473]
```

```python
# Update train and test dataframes with the new data

og_train_columns = train.columns.to_list()
og_test_columns = test.columns.to_list()
wrk = train.set_index('id')
for i in train_bad_addresses.columns.to_list():
    if i != 'id':
        wrk[i].update(train_bad_addresses.set_index('id')[i])

new_train = wrk.reset_index().reindex(columns=og_train_columns)

wrk = test.set_index('id')
for i in test_bad_addresses.columns.to_list():
    if i != 'id':
        wrk[i].update(test_bad_addresses.set_index('id')[i])

new_test = wrk.reset_index().reindex(columns=og_test_columns)
```

```python
new_train.to_csv('train_sber_corrected.csv', index=False)
new_test.to_csv('test_sber_corrected.csv', index=False)
```
